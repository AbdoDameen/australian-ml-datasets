## 0. Create Repository Folder Structure

Automatically create the folder hierarchy for organizing datasets by domain.

````python
import os
from pathlib import Path

# Define the base repository structure
base_path = Path('.')  # Current directory (australian-ml-datasets)

# Create main directories
main_dirs = ['datasets', 'notebooks', 'metadata', 'scripts', 'tests']

# Create subdirectories for datasets
dataset_domains = [
    'demographics',
    'weather',
    'housing',
    'crime',
    'transport',
    'health',
    'finance',
    'education',
    'environment',
    'elections',
    'business'
]

# Create subdirectories for notebooks
notebook_types = ['examples', 'preprocessing']

# Create subdirectories for scripts
script_types = ['data_cleaning']

# Function to create directories
def create_directory_structure():
    """Create the complete repository folder structure"""
    print("🏗️ Creating directory structure...\n")
    
    # Create main directories
    for main_dir in main_dirs:
        dir_path = base_path / main_dir
        dir_path.mkdir(exist_ok=True)
        print(f"✓ Created: {dir_path}")
    
    # Create dataset domain folders
    datasets_path = base_path / 'datasets'
    for domain in dataset_domains:
        domain_path = datasets_path / domain
        domain_path.mkdir(exist_ok=True)
        print(f"✓ Created: {domain_path}")
    
    # Create notebook subdirectories
    notebooks_path = base_path / 'notebooks'
    for notebook_type in notebook_types:
        notebook_path = notebooks_path / notebook_type
        notebook_path.mkdir(exist_ok=True)
        print(f"✓ Created: {notebook_path}")
    
    # Create script subdirectories
    scripts_path = base_path / 'scripts'
    for script_type in script_types:
        script_path = scripts_path / script_type
        script_path.mkdir(exist_ok=True)
        print(f"✓ Created: {script_path}")
    
    print("\n✅ Directory structure created successfully!")
    print(f"\n📂 Repository Structure:")
    print(f"australian-ml-datasets/")
    print(f"├── datasets/")
    for domain in dataset_domains:
        print(f"│   ├── {domain}/")
    print(f"├── notebooks/")
    for notebook_type in notebook_types:
        print(f"│   ├── {notebook_type}/")
    print(f"├── metadata/")
    print(f"├── scripts/")
    print(f"│   ├── data_cleaning/")
    print(f"└── tests/")

# Execute directory creation
create_directory_structure()
````

````python
def add_new_dataset(domain_name, dataset_name, create_metadata=True):
    """
    Add a new dataset by creating necessary folders and metadata files
    
    Parameters:
    -----------
    domain_name : str
        The domain folder (e.g., 'demographics', 'weather', 'housing')
    dataset_name : str
        The name of the specific dataset (e.g., 'census_2021', 'sydney_weather')
    create_metadata : bool
        Whether to create a metadata.json file
    """
    dataset_path = base_path / 'datasets' / domain_name / dataset_name
    dataset_path.mkdir(parents=True, exist_ok=True)
    
    # Create subdirectories for the dataset
    (dataset_path / 'raw').mkdir(exist_ok=True)
    (dataset_path / 'processed').mkdir(exist_ok=True)
    (dataset_path / 'features').mkdir(exist_ok=True)
    
    print(f"✅ Created dataset structure for '{dataset_name}' in '{domain_name}'")
    print(f"   ├── raw/          (original unprocessed data)")
    print(f"   ├── processed/    (cleaned data)")
    print(f"   └── features/     (engineered features)")
    
    # Create metadata file
    if create_metadata:
        metadata_file = dataset_path / 'metadata.json'
        metadata_content = {
            "dataset_name": dataset_name,
            "domain": domain_name,
            "created_date": str(datetime.now()),
            "description": f"Dataset for {domain_name}",
            "source": "To be filled",
            "license": "To be filled",
            "features": [],
            "rows": 0,
            "columns": 0
        }
        import json
        with open(metadata_file, 'w') as f:
            json.dump(metadata_content, f, indent=2)
        print(f"   ✓ Created: metadata.json")
    
    return dataset_path

# Example: Add new datasets
print("\n📦 Adding example datasets:\n")
add_new_dataset('demographics', 'abs_census_2021')
print()
add_new_dataset('weather', 'bom_sydney_2023')
print()
add_new_dataset('housing', 'nsw_prices_2023')
print()
add_new_dataset('transport', 'opal_trips_2023')
````

## 6. Data Cleaning and Preprocessing

Handle missing values, remove duplicates, standardize column names, convert data types, and address outliers across all datasets.

````python
# Function to clean and preprocess datasets
def preprocess_dataset(df, name="Dataset"):
    """
    Comprehensive data cleaning pipeline
    """
    print(f"\n🧹 Cleaning {name}...")
    
    # 1. Handle duplicates
    duplicates = df.duplicated().sum()
    df = df.drop_duplicates()
    print(f"   ✓ Removed {duplicates} duplicate rows")
    
    # 2. Standardize column names
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    print(f"   ✓ Standardized column names")
    
    # 3. Handle missing values
    missing_before = df.isnull().sum().sum()
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
    missing_after = df.isnull().sum().sum()
    print(f"   ✓ Handled {missing_before - missing_after} missing values")
    
    # 4. Remove outliers using IQR method
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        df = df[(df[col] >= Q1 - 1.5*IQR) & (df[col] <= Q3 + 1.5*IQR)]
    print(f"   ✓ Removed outliers (IQR method)")
    
    # 5. Convert data types
    df = df.convert_dtypes()
    print(f"   ✓ Optimized data types")
    
    return df

# Clean all datasets
demographics_clean = preprocess_dataset(demographics_df, "Demographics")
housing_clean = preprocess_dataset(housing_df, "Housing")
transport_clean = preprocess_dataset(transport_df, "Transport")

print(f"\n✅ All datasets cleaned!")
print(f"\nDemographics: {demographics_clean.shape}")
print(f"Housing: {housing_clean.shape}")
print(f"Transport: {transport_clean.shape}")
````

## 7. Feature Engineering

Create new features from existing data such as derived time-series features, aggregations, and domain-specific indicators.

````python
# Housing Feature Engineering
print("🏠 Housing Features:")
housing_features = housing_clean.copy()

# Create new features
housing_features['price_per_sqm'] = housing_features['price'] / housing_features['land_size_sqm']
housing_features['bedroom_to_bathroom_ratio'] = housing_features['bedrooms'] / (housing_features['bathrooms'] + 1)
housing_features['age'] = 2024 - housing_features['year_built']
housing_features['is_new'] = (housing_features['age'] <= 5).astype(int)

print(housing_features[['price', 'price_per_sqm', 'bedroom_to_bathroom_ratio', 'age']].head())

# Transport Feature Engineering
print("\n🚗 Transport Features:")
transport_features = transport_clean.copy()

# Temporal features
transport_features['is_peak_hour'] = transport_features['hour_of_day'].isin([7, 8, 9, 17, 18, 19]).astype(int)
transport_features['is_weekend'] = transport_features['day_of_week'].isin(['Saturday', 'Sunday']).astype(int)

# Aggregation features
transport_features['route_avg_passengers'] = transport_features.groupby('route')['passenger_count'].transform('mean')

print(transport_features[['hour_of_day', 'is_peak_hour', 'is_weekend', 'route_avg_passengers']].head())

print("\n✅ Feature engineering complete!")
````

## 8. Create Train/Test Splits

Split datasets into training and testing subsets using scikit-learn, maintaining temporal order for time-series data.

````python
# Standard train/test split for housing data
X_housing = housing_features[['bedrooms', 'bathrooms', 'land_size_sqm', 'age', 'price_per_sqm']]
y_housing = housing_features['price']

X_train, X_test, y_train, y_test = train_test_split(
    X_housing, y_housing, test_size=0.2, random_state=42
)

print("🏠 Housing Data Split:")
print(f"   Training set: {X_train.shape[0]} samples")
print(f"   Testing set: {X_test.shape[0]} samples")
print(f"   Features: {X_train.shape[1]}")

# Temporal train/test split for transport data (time-series)
split_point = int(len(transport_features) * 0.8)
transport_train = transport_features.iloc[:split_point]
transport_test = transport_features.iloc[split_point:]

print(f"\n🚗 Transport Data Split (Temporal):")
print(f"   Training set: {len(transport_train)} samples")
print(f"   Testing set: {len(transport_test)} samples")
print(f"   Train date range: {transport_train['timestamp'].min()} to {transport_train['timestamp'].max()}")
print(f"   Test date range: {transport_test['timestamp'].min()} to {transport_test['timestamp'].max()}")

# Standardize numeric features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"\n✅ Train/Test split and scaling complete!")
print(f"   Scaled training shape: {X_train_scaled.shape}")
print(f"   Scaled testing shape: {X_test_scaled.shape}")
````
